```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_openml
```

```python
# Descargar el dataset "Zoo"
zoo_dataset = fetch_openml(name='zoo', version=1)
zoo_df = pd.DataFrame(data=zoo_dataset.data, columns=zoo_dataset.feature_names)
```

```python
print(zoo_df.columns)
print(zoo_df.head())


# Gráfica de barras para la variable 'hair'
plt.figure(figsize=(10, 6))
sns.countplot(x='hair', data=zoo_df)
plt.title('Cantidad de animales con pelo')
plt.xlabel('¿Tiene pelo? (0: No, 1: Sí)')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

# Gráfica de barras para la variable 'feathers'
plt.figure(figsize=(10, 6))
sns.countplot(x='feathers', data=zoo_df)
plt.title('Cantidad de animales con plumas')
plt.xlabel('¿Tiene plumas? (0: No, 1: Sí)')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()

# Gráfica de barras para la variable 'aquatic'
plt.figure(figsize=(10, 6))
sns.countplot(x='aquatic', data=zoo_df)
plt.title('Cantidad de animales acuáticos')
plt.xlabel('¿Es acuático? (0: No, 1: Sí)')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
plt.show()
```

```python
# Convertir columnas a tipo numérico
for col in ['hair', 'feathers', 'eggs', 'milk', 'airborne', 
            'aquatic', 'predator', 'toothed', 'backbone', 
            'breathes', 'venomous', 'fins', 'legs', 'tail', 
            'domestic', 'catsize']:
    zoo_df[col] = pd.to_numeric(zoo_df[col], errors='coerce')

    
# Obtener parámetros estadísticos
stats = zoo_df.describe()
print(stats)
```
